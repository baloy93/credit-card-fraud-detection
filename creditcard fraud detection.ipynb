```python
from azure.eventhub import EventHubProducerClient, EventData
import json
import random
import time

connection_string = "Endpoint=sb://creditcard-transactions-eventhub.servicebus.windows.net/;SharedAccessKeyName=RootManageSharedAccessKey;SharedAccessKey=BNlOQILAO0MJ78PprVAG2dPaeqTGQEsg9+AEhPju+JM="
eventhub_name = "creditcardfraud "

# Creating a producer client
producer = EventHubProducerClient.from_connection_string(connection_string, eventhub_name=eventhub_name)

try:
    while True:
        # Simulating transaction data
        transaction = {
            "TransactionId": random.randint(1, 10000),
            "Amount": round(random.uniform(5, 15000), 2),
            "Location": random.choice(["LocationA", "LocationB", "LocationC"]),
            "Timestamp": time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())
        }

        # Creating an event
        event_data = EventData(json.dumps(transaction))

        # Sending the event to the Event Hub
        with producer:
            producer.send_event(event_data)

        print(f"Sent: {transaction}")
        time.sleep(1)  # Sleep for a second before sending the next transaction

except KeyboardInterrupt:
    print("Stopped sending events.")
finally:
    producer.close()
```
